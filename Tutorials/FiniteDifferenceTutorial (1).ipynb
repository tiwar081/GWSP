```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# True functions
def f(x):
    return np.cos(x**2+x+3)
def fprime(x):
    return -(2*x+1)*np.sin(x**2+x+3)

# Arrays for plotting exact results
x = np.linspace(-5, 5, 1000)
y = f(x)
dydx = fprime(x)
```

```python
plt.plot(x, y, label = r'$f(x)=\cos(x^2+x+3)$')
plt.plot(x, dydx, label = r'$f\,^\prime(x)=-(2x+1)\sin(x^2+x+3)$')
plt.title('True function and derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (20,10))
deltax = np.array([1, 0.5, 0.1, 0.01])
dydx_fd = np.zeros([5, 1000])
for i in range(deltax.shape[0]):
    dydx_fd[i] = (f(x + deltax[i]) - f(x))/deltax[i]
    plt.plot(x, dydx_fd[i], label = 'finite difference derivative for ' + r'$\delta x =$' + str(deltax[i]))
plt.plot(x, dydx, label = 'exact derivative', linestyle='dashed', color = 'black')
plt.title('Finite difference derivative approximations', fontsize = 30)
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize = 20)
plt.show()
```

```python
targetx = 3 # arbitrary
targetfprime = fprime(targetx)
targetfprime_arr = np.full(1000, targetfprime)
deltax_arr = np.geomspace(1e-20, 1, 1000)
dydx_fd_2 = (f(targetx + deltax_arr) - f(targetx))/deltax_arr
plt.figure(figsize = (20,10))
plt.semilogx(deltax_arr, dydx_fd_2, label = 'finite difference derivative')
plt.semilogx(deltax_arr, targetfprime_arr, label = 'exact derivative', linestyle='dashed', color = 'black')
plt.title('Accuracy of finite differencing at x = ' + str(targetx), fontsize = 30)
plt.xlim(1, 1e-20) #reverse order to make the x axis backwards
plt.xlabel(r'$\delta x$',fontsize = 20)
plt.ylabel(r'$f\,^\prime(x)$',fontsize = 20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize = 20)
plt.show()
```
